```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path = "results/CartPole-v0_dqn_10_22_23_19.txt"
```

```python
results = pd.read_table(path, delimiter=",", names=['episode','score'])
```

```python
results['roll'] = results['score'].rolling(3).mean()
```

```python
fig,ax = plt.subplots(2,1,figsize=(20,5), sharex=True)
ax[0].plot(results['episode'],results['score'])
ax[1].plot(results['episode'],results['roll'])

ax[1].set_xlabel("episodes")
ax[0].set_ylabel("reward")
ax[1].set_ylabel("reward")
_ = ax[0].set_title("reward curb for Deep Q Learning")
_ = ax[1].set_title("moving avg reward curb for Deep Q Learning")
```
